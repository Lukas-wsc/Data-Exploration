# Preprocessing twitter-corpus-data that is used to validate the models

importing all necessary libraries

```python
import pandas as pd # for data processing
```

setting to show full text content

```python
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('full-corpus.csv', header=None)
```

```python
df
```

dropping unnecessary columns, only keeping 1 (as label) and 4 (as feature)

```python
df = df.drop([0, 2, 3], axis =1)
```

```python
df
```

```python
new_header = ['Ground_Truth_Label', 'Text']
df.columns = new_header
```

only keeping positive and negative, dropping irrelevant and neutral

```python
df_filtered = df.loc[df['Ground_Truth_Label'].isin(['positive', 'negative'])]
```

```python
df_filtered
```

downsampling classes to equal size

```python
value_counts = df_filtered['Ground_Truth_Label'].value_counts()
print(value_counts)
```

```python
minority_class = value_counts.idxmin()
majority_class = value_counts.idxmax()
```

```python
minority_count = value_counts.min()
print(minority_count)
```

```python
df_majority = df[df['Ground_Truth_Label'] == majority_class]
df_minority = df[df['Ground_Truth_Label'] == minority_class]
```

```python
df_majority
```

```python
df_minority
```

```python
df_majority_downsampled = df_majority.sample(n=minority_count, random_state=42)
```

```python
df_balanced = pd.concat([df_minority, df_majority_downsampled])
```

```python
print("\nValue Counts after balancing:")
print(df_balanced['Ground_Truth_Label'].value_counts())
```

adding target number to dataset

```python
def map_target_to_label(target):
    if target == 'positive':
        return 4
    elif target == 'negative':
        return 0
    else:
        return None
df_balanced['target_number'] = df_balanced['Ground_Truth_Label'].apply(map_target_to_label)
```

```python
df_balanced
```

exporting dataset

```python
df_balanced.to_csv('full_corpus_preprocessed.csv', index=False)
```
