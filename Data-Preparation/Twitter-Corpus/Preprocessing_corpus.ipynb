```python
minority_count = value_counts.min()
print(minority_count)
```

```python
# DataFrames für jede Klasse
df_majority = df[df['target'] == majority_class]
df_minority = df[df['target'] == minority_class]

```

```python
df_majority
```

```python
df_minority
```

```python
# Downsampling der Mehrheitklasse
df_majority_downsampled = df_majority.sample(n=minority_count, random_state=42)
```

```python
# Zusammenführen der ausgeglichenen DataFrames
df_balanced = pd.concat([df_minority, df_majority_downsampled])
```

```python
print("\nValue Counts after balancing:")
print(df_balanced['target'].value_counts())
```

## Stopword-cleaning

```python
def clean_text(text):
    stop_words = stopwords.words('english')
    text = re.sub(r'[@#]\w+', ' ', text)  # Entfernen von Hashtags und Verlinkungen 
    text = re.sub(r'https?://\S+', ' ', text)  # Entfernen von Links
    text = re.sub(r"\b\w+['’]\w+\b", ' ', text)  # Entfernen von Kontraktionen wie 's, 'm etc. 
    text = re.sub(rf"[{punctuation}]", ' ', text)  # Entfernen von Satzzeichen
    words = text.split()  
    words = [word for word in words if word.lower() not in stop_words]  
    return ' '.join(words)
```
